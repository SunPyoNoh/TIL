```python
import re

file = open('./young_king.txt','r',encoding='utf-8')
lines = file.read()

sentences = re.split('(?<=[2-9]\n)|(?<=[0-9][0-9]\n)|(?<=[0-9][0-9][0-9]\n)', lines)

sent=[]
for stuff in sentences:
        sent.append(stuff)
```

```python
print(sent)
```

```python
twitter = Twitter()
```

```python
sentences_tag = []
for sentence in sent:
    morph = twitter.pos(sentence)
    sentences_tag.append(morph)
```

```python
print(sentences_tag)
```

```python
noun_adj_list = []
noun_adj=[]

for sentence1 in sentences_tag:
    for word, tag in sentence1:
        if tag in ['Noun']:
            noun_adj.append(word)
```

```python
import pandas as pd
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.manifold import TSNE
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vector = CountVectorizer()
print(vector.fit_transform(noun_adj).toarray()) # 코퍼스로부터 각 단어의 빈도 수를 기록한다.
print(vector.vocabulary_) # 각 단어의 인덱스가 어떻게 부여되었는지를 보여준다.
```

```python
tfidfv = TfidfVectorizer().fit(noun_adj)
print(tfidfv.transform(noun_adj).toarray())
print(tfidfv.vocabulary_)
```
